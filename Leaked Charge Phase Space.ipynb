```python
import matplotlib.pyplot as plt
import numpy as np
```

Small m regime, taylor expand, epsilon = 10^-10/m

```python
m = np.logspace(-10,-1,num=10000)
epsilon=10**(-10)/m
```

```python
plt.plot(m,epsilon)
plt.xscale("log")
plt.yscale("log")
```

```python
a=0.5*10**(-3)
q=0.3
m = np.logspace(-4,5,num=1000)
```

```python
small_a_factor = 1-np.sinh(m*a)/(m*a)
```

```python
epsilon = 10**(-14)/np.sqrt(np.abs(small_a_factor*q))
```

```python
im = plt.imread("DP phase space 2013.png")
implot = plt.imshow(im,extent=[10**-20,10**6,10**0,10**-23])
plt.plot(m,epsilon)
plt.xscale("log")
plt.yscale("log")
plt.xlim(10**-20,10**6)
plt.ylim(10**-23,1)
plt.show()
```
