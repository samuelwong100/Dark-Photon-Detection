```python
import numpy as np
%pylab inline
from scipy.special import shichi
```

```python
m = np.logspace(-5,-1)
```

# Plane Model

```python
def epsilon(m,x,entanglement=True):
    if entanglement:
        power = 10**(-9)
    else:
        power = 10**(-8)
    return np.sqrt(3)*power*(m**(-0.5))*np.exp(m*x/2)
```

```python
plt.plot(m,epsilon(m,0.5*10**5,False),label="1 cm",linestyle="--",color = "grey")
plt.plot(m,epsilon(m,0.5*10**5,True),label="1 cm, entanglement",linestyle="-",color = "grey")
plt.plot(m,epsilon(m,0.5*10**4,False),label="1 mm",linestyle="--",color = "green")
plt.plot(m,epsilon(m,0.5*10**4,True),label="1 mm, entanglement",linestyle="-",color = "green")
plt.plot(m,epsilon(m,0.5*10**3,False),label="100 micron",linestyle="--",color = "blue")
plt.plot(m,epsilon(m,0.5*10**3,True),label="100 micron, entanglement",linestyle="-",color = "blue")
plt.plot(m,epsilon(m,0.5*10**2,False),label="10 micron",linestyle="--",color = "red")
plt.plot(m,epsilon(m,0.5*10**2,True),label="10 micron, entanglement",linestyle="-",color = "red")
plt.plot(m,10**(-32/3)*m**(-2/3),color="black",label="stellar")
plt.xscale('log')
plt.yscale('log')
plt.ylim(10**-10,10**-4)
plt.xlabel("m [eV]")
plt.ylabel("$\epsilon$")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title("Neutrality Test (Plane Model)")
plt.savefig("Neutrality Test (Plane Model).png", dpi=500, bbox_inches='tight')
```

# Sphere Model

```python
def h(m,R,b):
    r = R-b
    summation = 10*np.sinh(m*r)/(m*r) - 7*shichi(m*r)[1] + 7*np.cosh(m*r)/(m*r)**2 + 2*np.sinh(m*r)/(m*r)**3
    return np.exp(m*R/2)*summation**(-1/2)

def epsilon_sphere(m,b,R,entanglement=True):
    if entanglement:
        power = 10**(-11)
    else:
        power = 10**(-10)
    return np.sqrt(2)*power*h(m,R,b)/m
```

```python
def plot_sphere(R,string):
    if R > 10**5:
        plt.plot(m,epsilon_sphere(m,5*10**4,R,False),label="1 cm",linestyle="--",color = "grey")
        plt.plot(m,epsilon_sphere(m,5*10**4,R,True),label="1 cm, entanglement",linestyle="-",color = "grey")
    if R > 10**4:
        plt.plot(m,epsilon_sphere(m,5*10**3,R,False),label="1 mm",linestyle="--",color = "green")
        plt.plot(m,epsilon_sphere(m,5*10**3,R,True),label="1 mm, entanglement",linestyle="-",color = "green")
    plt.plot(m,epsilon_sphere(m,5*10**2,R,False),label="100 micron",linestyle="--",color = "blue")
    plt.plot(m,epsilon_sphere(m,5*10**2,R,True),label="100 micron, entanglement",linestyle="-",color = "blue")
    plt.plot(m,epsilon_sphere(m,5*10**1,R,False),label="10 micron",linestyle="--",color = "red")
    plt.plot(m,epsilon_sphere(m,5*10**1,R,True),label="10 micron, entanglement",linestyle="-",color = "red")
    plt.plot(m,10**(-32/3)*m**(-2/3),color="black",label="stellar")
    plt.xscale('log')
    plt.yscale('log')
    plt.ylim(10**-10,10**-4)
    plt.xlabel("m [eV]")
    plt.ylabel("$\epsilon$")
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.title("Neutrality Test (Sphere Model, R = {})".format(string))
    plt.savefig("Neutrality Test (Sphere Model, R = {}).png".format(string), dpi=500, bbox_inches='tight')
```

```python
R = 2 * 10**(5) # 4.5 cm
plot_sphere(R, "4.5 cm")
```

```python
R = 2 * 10**(4) # 4.5 mm
plot_sphere(R, "4.5 mm")
```

```python
R = 2 * 10**(3) # 0.45 mm
plot_sphere(R, "0.45 mm")
```
